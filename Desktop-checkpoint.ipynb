```python
def time_rest(x):
    if x.isalpha():
        pass
    elif x=="Early morning" or "Early morning":
        x="Morning"
    elif "evening" in x.lower():
        x='Evening'
    elif x=="overnight":
        x="Overnight"
    else:
        if int(x.split(':')[0])>12 and int(x.split(":")[0])<18:
            x="Afternoon"
        elif int(x.split(':')[0])<=12:
            x="Morning"
        else:
            x="Night"
    return x
```

```python
a['time']=a['time'].apply(time_rest)
```
